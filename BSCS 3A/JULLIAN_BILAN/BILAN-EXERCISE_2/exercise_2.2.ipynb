## Part A: Extract Words Starting with Uppercase (10 points)
Extract all words starting with an uppercase letter from the given text.

**RegEx Pattern:** `[A-Z][a-z]*`

```python
# Part A: Extract words starting with uppercase letter

text = '''Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do. Once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, "and what is the use of a book," thought Alice, "without pictures or conversations?'''

# RegEx pattern: [A-Z][a-z]* - Matches words starting with uppercase followed by lowercase letters
pattern = r'[A-Z][a-z]*'

# Find all matches
uppercase_words = re.findall(pattern, text)

print("Words starting with uppercase letter:")
print(uppercase_words)
print(f"\nTotal count: {len(uppercase_words)}")
```

## Part B: Extract and Replace "Whale" Instances (15 points)
Read "melville-moby_dick.txt" and extract all instances of Whale, Whales, whale, whales. Then replace the first 10 instances with "leviathan".

**RegEx Pattern:** `[Ww]hales?`

```python
# Part B: Read Moby Dick and extract/replace whale instances

# Read the file
with open('melville-moby_dick.txt', 'r', encoding='utf-8') as file:
    moby_dick_text = file.read()

# RegEx pattern: [Ww]hales? - Matches Whale, whale, Whales, whales
pattern = r'[Ww]hales?'

# Find all instances
whale_instances = re.findall(pattern, moby_dick_text)

print("All instances of Whale/whale/Whales/whales found:")
print(whale_instances[:20])  # Show first 20 for preview
print(f"\nTotal instances found: {len(whale_instances)}")
```

```python
# Replace first 10 instances of whale/Whale/whales/Whales with "leviathan"

# Counter for replacements
replacement_count = [0]

def replace_whale(match):
    """Replace function that only replaces first 10 instances"""
    if replacement_count[0] < 10:
        replacement_count[0] += 1
        return "leviathan"
    return match.group(0)

# Perform the replacement
modified_text = re.sub(pattern, replace_whale, moby_dick_text)

# Verify by finding first 15 instances of leviathan or whale pattern
verification_pattern = r'leviathan|[Ww]hales?'
first_instances = re.findall(verification_pattern, modified_text)

print("First 15 instances after replacement (showing leviathan replacements):")
print(first_instances[:15])
print(f"\nTotal replacements made: {replacement_count[0]}")
```

## Part C: Extract Jack Sparrow's Lines (15 points)
Download NLTK, import webtext, load pirates.txt, and extract all lines spoken by Jack Sparrow.

**RegEx Pattern:** `JACK SPARROW: (.+)`

```python
# Part C: Install and import NLTK (run this cell first)
import nltk
nltk.download('webtext')
```

```python
# Part C: Extract Jack Sparrow's lines from pirates.txt

from nltk.corpus import webtext

# Load pirates.txt
pirates_text = webtext.raw('pirates.txt')

# RegEx pattern: JACK SPARROW: (.+) - Captures everything after "JACK SPARROW: "
pattern = r'JACK SPARROW: (.+)'

# Find all lines spoken by Jack Sparrow
jack_sparrow_lines = re.findall(pattern, pirates_text)

print("Lines spoken by Jack Sparrow:")
print("-" * 50)
for i, line in enumerate(jack_sparrow_lines, 1):
    print(f"{i}. {line}")
    
print("-" * 50)
print(f"\nTotal lines spoken by Jack Sparrow: {len(jack_sparrow_lines)}")
```
